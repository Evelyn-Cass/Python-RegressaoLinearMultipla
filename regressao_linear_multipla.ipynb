```python
import numpy as np
```

```python
# 1.Lendo os dados com NumPy
# Pulamos o cabeçalho (skip_header =1)
# Extraímos apenas as colunas de interesse para evitar erro com strings:
# 2: Price (y), 3: bedrooms (X1), 4: bathrooms (X2), 5: sqft_living (X3)
data = np.genfromtxt(
    "kc_house_data.csv", delimiter=",", skip_header=1, usecols=(2, 3, 4, 5)
)
```

```python
# 2. Separando a variável alvo (y) e as variáveis preditoras (X)
y = data[:, 0]  # O preço é a primeira coluna do nosso recorte
X_features = data[:, 1:]  # O restante são features
```

```python
# 3. Acionando o termo de Intercepto (Viés/Bias)
# Precisamos adicionar uma coluna de números "1" no início da matriz X
# Isso representa o beta_0 na equação ( o valor base na casa se tudo fosse zero)
m = X_features.shape[0]  # Número de linhas (exemplos)
X = np.c_[np.ones(m), X_features]
```

```python

# 4. Aplicando a Equação Normal: Beta = (X.T * X)^(-1) * X.T * y
X_T = X.T  # Transposta de X
# np.linalg.inv calcula a inversa da matriz
# o método .dot() ou o operador @ realiza a multiplicação de matrizes
# beta = np.linalg.inv(X_T.dot(X)).dot(X_T).dot(y)
beta = np.linalg.inv(X_T @ X) @ X_T @ y
```

```python
print("--- Coeficientes Descobertos ---")
print(f"Intercepto (Beta 0): {beta[0]:.2f}")
print(f"Peso para Quartos (Beta 1): {beta[1]:.2f}")
print(f"Peso para Banheiros (Beta 2): {beta[2]:.2f}")
print(f"Peso para Área/sqft (Beta 3): {beta[3]:.2f}")
```

```python
#5. Fazendo uma previsão
# Vamos prever o valor da primeira casa do dataset: 3 quartos, 1 banheiro, 1180 sqft
casa_teste = np.array([1,3,1,1180]) # Lembre-se do "1" inicial para o intercepto! 
preco_previsto = casa_teste @ beta

print(f"--- Teste de Previsão ---")
print(f"Preço REAL da casa 1: ${y[0]:.2f}")
print(f"Preço PREVISTO pelo nosso modelo: ${preco_previsto:.2f}")
```
